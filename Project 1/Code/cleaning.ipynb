```python
#importing dependecies
import pandas as pd 
import numpy as np 

#Reading data into dataframe
sports_data_df = pd.read_csv("../Resources/nfl_offensive_stats.csv")
```

```python
#Showing Data frame
sports_data_df
```

```python
#Creating new df with columns useful for determining correlation between weather and total offensive yards
weather_yards_df = sports_data_df[["game_id", "player", "team", "pass_yds", "rush_yds", "Temperature", "Humidity", "Wind_Speed", "game_date"]].copy()
weather_yards_df
```

```python
#Grouping by unique game to see total rush and pass yards per game
weather_yards_df1 = weather_yards_df.groupby("game_id")[["pass_yds", "rush_yds"]].sum()
weather_yards_df1
```

```python
#Grouping to see the temperature and wind speed per unique game
weather_yards_df2 = weather_yards_df.groupby("game_id")[["Temperature", "Wind_Speed"]].first()
weather_yards_df2
```

```python
#Merging the two df's together to make one
weather_yards_df3 = pd.merge(weather_yards_df1, weather_yards_df2, how="left", on = "game_id")
weather_yards_df3
```

```python
#Creating bins to make categories for different temperatures
temperature_bins = [0, 32, 60, 90, 120]
bin_labels = ["0 - 32 Degrees", "32 - 60 Degrees", "60 - 90 Degrees", "90 - 120 Degrees"]
final_weather_yards_df = weather_yards_df3.copy()
final_weather_yards_df
```

```python
#Adding the bins into new column called temperature range
final_weather_yards_df["Temperature Range"] = pd.cut(x=final_weather_yards_df["Temperature"], bins=temperature_bins, labels=bin_labels)
final_weather_yards_df
```

```python
#Sorting by temp
final_weather_yards_df.sort_values(by="Temperature")
```

```python
#grouping by temp range and the mean of pass and rush yards
ready_to_graph_temp_vs_yards_df = final_weather_yards_df.groupby("Temperature Range")[["pass_yds", "rush_yds"]].mean()
ready_to_graph_temp_vs_yards_df
```

```python
#Creating bins for wind speed
final_wind_yards_df = weather_yards_df3.copy()
wind_speed_bins = [-1, 5, 10, 15, 20, 25, 30, 40]
wind_speed_labels = ["0 - 5", "5 - 10", "10 - 15", "15 - 20", "20 - 25", "25 - 30", "30 - 40"]
final_wind_yards_df.sort_values(by="Wind_Speed")
```

```python
#Adding wind speed range column
final_wind_yards_df["Wind Speed Range (mph)"] = pd.cut(x=final_wind_yards_df["Wind_Speed"], bins=wind_speed_bins, labels=wind_speed_labels)
final_wind_yards_df.sort_values(by="Wind_Speed")
```

```python
#Grouping by wind speed range and taking the average of pass yards and rush yards
ready_to_graph_wind_vs_yards_df = final_wind_yards_df.groupby("Wind Speed Range (mph)")[["pass_yds", "rush_yds"]].mean()
ready_to_graph_wind_vs_yards_df
```

```python
#exporting the two data frames to be used in graphing relating weather and yards
ready_to_graph_temp_vs_yards_df.to_csv("../Output/Temp_vs_Yards.csv")
ready_to_graph_wind_vs_yards_df.to_csv("../Output/Wind_vs_Yards.csv")
```

```python
#Creating new table with columns to see home vs away yard and score difference
home_away_df = sports_data_df[["game_id", "player", "team", "pass_yds", "rush_yds", "home_team", "vis_team", "home_score", "vis_score"]].copy()
home_away_df
```

```python
#Grouping by the game and the team and getting pass and rush yards
home_away_df1 = home_away_df.groupby(["game_id", "team"])[["pass_yds", "rush_yds"]].sum()
home_away_df1 
```

```python
#grouping by game and team and getting home and away team and score
home_away_df2 = home_away_df.groupby(["game_id", "team"])[["home_team", "vis_team", "home_score", "vis_score"]].first()
home_away_df2
```

```python
#merging the two dataframes together into one
home_away_df3 = pd.merge(home_away_df1, home_away_df2, how="left", on = ["game_id", "team"])
home_away_df3
```

```python
#Declaring variables for the home team
count = 0
homePass1 = ""
homePass = []
homeRush1 = ""
homeRush = []
homeScore1 = ""
homeScore = []

#Declaring variables for the away team
awayPass1 = ""
awayPass = []
awayRush1 = ""
awayRush = []
awayScore = ""
awayScore = []

#Creating a loop to go through the dataframe
for index, row in home_away_df3.iterrows():
    
    #creating if to get the home vs away team and store their values
    if(count == 0):

        homePass1 = row["pass_yds"]
        homePass.append(homePass1)
        homeRush1 = row["rush_yds"]
        homeRush.append(homeRush1)
        homeScore1 = row["home_score"]
        homeScore.append(homeScore1)

        count = count + 1
    
    #home team data being aquired
    elif(count % 2 == 0):

        count = count + 1

        homePass1 = row["pass_yds"]
        homePass.append(homePass1)
        homeRush1 = row["rush_yds"]
        homeRush.append(homeRush1)
        homeScore1 = row["home_score"]
        homeScore.append(homeScore1)
    
    #away team data being aquired
    else:
        count = count + 1

        awayPass1 = row["pass_yds"]
        awayPass.append(awayPass1)
        awayRush1 = row["rush_yds"]
        awayRush.append(awayRush1)
        awayScore1 = row["vis_score"]
        awayScore.append(awayScore1)

#Creating data frames with data from home teams and another one for away teams   
home_df = pd.DataFrame({"Pass Yards" : homePass, "Rush Yards" : homeRush, "Score" : homeScore})
away_df = pd.DataFrame({"Pass Yards" : awayPass, "Rush Yards" : awayRush, "Score" : awayScore})   

#showing home team table
home_df
```

```python
#Getting the avg of pass and rush yards and score for home teams
homeAvgPass = home_df["Pass Yards"].mean()
homeAvgRun = home_df["Rush Yards"].mean()
homeAvgScore = home_df["Score"].mean()

#creating a table of the avg pass yards, rush yards, and score for home teams
homeAvg_df = pd.DataFrame({"Pass Yards" : homeAvgPass, "Rush Yards" : homeAvgRun, "Score" : homeAvgScore}, index=[0])
homeAvg_df
```

```python
#Getting the avg of pass and rush yards and score for away teams
awayAvgPass = away_df["Pass Yards"].mean()
awayAvgRun = away_df["Rush Yards"].mean()
awayAvgScore = away_df["Score"].mean()

#creating a table of the avg pass yards, rush yards, and score for away teams
awayAvg_df = pd.DataFrame({"Pass Yards" : awayAvgPass, "Rush Yards" : awayAvgRun, "Score" : awayAvgScore}, index=[0])
awayAvg_df
```

```python
#Adding a home or away catergory
homeAvg_df["Home or Away"] = "Home"
awayAvg_df["Home or Away"] = "Away"
```

```python
#Grouping by home or away
homeAvg_df = homeAvg_df.groupby(["Home or Away"])[["Pass Yards", "Rush Yards", "Score"]].first()
awayAvg_df = awayAvg_df.groupby(["Home or Away"])[["Pass Yards", "Rush Yards", "Score"]].first()
```

```python
#Merging home and away df
home_away_merged_df = pd.merge(homeAvg_df, awayAvg_df, how="outer", on=["Home or Away", "Pass Yards", "Rush Yards", "Score"])
home_away_merged_df
```
